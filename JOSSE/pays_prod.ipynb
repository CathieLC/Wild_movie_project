```python
import pandas as pd
```

```python
basics_df = pd.read_csv("https://datasets.imdbws.com/title.basics.tsv.gz", sep="\t", low_memory=False)
```

```python
akas_df = pd.read_csv("https://datasets.imdbws.com/title.akas.tsv.gz", sep="\t", dtype={'isOriginalTitle': str})
```

```python
#on isole les films en filtrant sur la colonne titleType
basics_df_movies =  basics_df[basics_df['titleType'] == 'movie']
```

```python
#on remplace les \N par de NaT
basics_df_movies_nan = basics_df_movies.replace('\\N', pd.NaT)
```

```python
#on récupère les colonnes tconst et runtimeMinutes
movies = basics_df_movies_nan[['tconst', 'runtimeMinutes','titleType']]
```

```python
#on retire les lignes dans lesquelles on a un NaT
movies_dropna = movies.dropna()
```

```python
#on transforme le runtimeMinutes en unt
movies_int = movies_dropna.astype({'runtimeMinutes': int})
```

```python
# on séléctionne les long métrage de plus de 58 minutes et de moins de 4h30
movies_int = movies_int[(movies_int['runtimeMinutes'] >= 58) & (movies_int['runtimeMinutes'] <= 270)]
```

```python
#jointure entre la table title basics filtrée et la table title akas
film_list = pd.merge(movies_int,akas_df,how='left',left_index=True,right_index=True)
```

```python
#recherche des long métrage en vf
list_movies_vfr = film_list[film_list['language'] == 'fr']
list_movies_vfr.tail(10)
```
